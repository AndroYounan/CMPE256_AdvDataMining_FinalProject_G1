```python
# Function to load JSON file into Pandas DataFrame
def load_json(filename):
    """Loads a JSON file into a pandas DataFrame"""
    data = []
    with open(filename, 'r', encoding='utf-8') as f:
        for line in f:
            data.append(json.loads(line))
    return pd.DataFrame(data)

andrew_path = r"C:\BIANCONERI\Master's AI SJSU\5- Advanced Data Mining\Final Project\Yelp-JSON\Yelp JSON\yelp_dataset"
nic_path = r"C:\Users\nicor\OneDrive\Documents\SJSU\CMPE256\group_proj\Yelp JSON"

review_df = load_json(nic_path + r"\yelp_academic_dataset_review.json")
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import datetime as dt

# Generate sample date data (replace this with your actual data)
np.random.seed(42)
start_date = dt.datetime(2023, 1, 1)
end_date = dt.datetime(2023, 12, 31)
days_range = (end_date - start_date).days

# Create random dates with more points in some months than others
random_days = np.random.randint(0, days_range, size=500)
dates = [start_date + dt.timedelta(days=float(day)) for day in sorted(random_days)]

# Convert to pandas DataFrame
df = pd.DataFrame({'date': dates})

# Extract month information and count points per month
df['year_month'] = df['date'].dt.strftime('%Y-%m')

# Group by month and count
monthly_counts = df.groupby('year_month').size().reset_index(name='count')

# Generate all months in range (in case some months have zero data points)
all_months = pd.date_range(start=start_date, end=end_date, freq='MS')
all_months_df = pd.DataFrame({
    'year_month': all_months.strftime('%Y-%m')
})

# Merge to ensure all months are represented
monthly_counts = pd.merge(all_months_df, monthly_counts, on='year_month', how='left').fillna(0)

# Calculate cumulative sum
monthly_counts['cumulative_count'] = monthly_counts['count'].cumsum()

# Plotting
plt.figure(figsize=(12, 6))

# Plot cumulative line
plt.plot(monthly_counts['year_month'], monthly_counts['cumulative_count'], 
         marker='o', linestyle='-', linewidth=2, color='#3366cc')

# Plot monthly count as bars
plt.bar(monthly_counts['year_month'], monthly_counts['count'], alpha=0.3, color='#3366cc')

# Customize the plot
plt.title('Cumulative Number of Data Points Per Month', fontsize=16)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Number of Data Points', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(rotation=45)
plt.tight_layout()

# Add annotations
for i, row in monthly_counts.iterrows():
    plt.annotate(f"{int(row['cumulative_count'])}",
                 xy=(i, row['cumulative_count']),
                 xytext=(0, 10), 
                 textcoords='offset points',
                 ha='center',
                 fontsize=9)

# Add a second y-axis for monthly counts
ax2 = plt.gca().twinx()
ax2.set_ylabel('Monthly Count', color='#3366cc', fontsize=12)
ax2.tick_params(axis='y', labelcolor='#3366cc')

# Add legend
plt.legend(['Cumulative Count', 'Monthly Count'], loc='upper left')

# Show the plot
plt.show()

# Print summary table
print("Monthly and Cumulative Data Points:")
print(monthly_counts)
```
