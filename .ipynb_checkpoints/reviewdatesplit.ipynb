```python
import pandas as pd
import json
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Function to load JSON file into Pandas DataFrame
def load_json(filename):
    """Loads a JSON file into a pandas DataFrame"""
    data = []
    with open(filename, 'r', encoding='utf-8') as f:
        for line in f:
            data.append(json.loads(line))
    return pd.DataFrame(data)

andrew_path = r"C:\BIANCONERI\Master's AI SJSU\5- Advanced Data Mining\Final Project\Yelp-JSON\Yelp JSON\yelp_dataset"
nic_path = r"C:\Users\nicor\OneDrive\Documents\SJSU\CMPE256\group_proj\Yelp JSON"

review_df = load_json(nic_path + r"\yelp_academic_dataset_review.json")
business_df = load_json(nic_path + r"\yelp_academic_dataset_business.json")
business_df = business_df[(business_df['state'] == 'CA') & (business_df['is_open'] == 1) & (business_df['categories'].notna())]
```

```python
review_df = review_df[review_df['business_id'].isin(business_df['business_id'])]
```

```python
print(review_df['date'].min())
print(review_df['date'].max())
review_df['yyyy-mm'] = review_df['date'].str[0:7]
monthlies = pd.DataFrame()
monthlies['counts'] = review_df.groupby('yyyy-mm').size()
monthlies['cumulative'] = monthlies['counts'].cumsum()
# monthlies['year-mo'] = review_df['date'].str[0:7]

```

```python
monthlies.head()
monthlies.tail()
total = monthlies['cumulative'].iloc[-1]
splitpt = total * 0.8
print(splitpt)
monthlies['splitabs'] = abs(monthlies['cumulative']-splitpt)
splitmth = monthlies['splitabs'].argmin()
monthlies.iloc[splitmth]
```

```python
_ = plt.figure(figsize=(20,6))
_ = plt.plot(monthlies['counts'])
_ = plt.xticks(rotation=90)
_ = plt.axvline(x=splitmth,color='black')
```

```python
_ = plt.figure(figsize=(20,6))
_ = plt.plot(monthlies['cumulative'])
_ = plt.xticks(rotation=90)
_ = plt.axvline(x=splitmth,color='black')
_ = plt.axhline(y = splitpt, color='red')
```
