```python
import pandas as pd
import json
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Function to load JSON file into Pandas DataFrame
def load_json(filename):
    """Loads a JSON file into a pandas DataFrame"""
    data = []
    with open(filename, 'r', encoding='utf-8') as f:
        for line in f:
            data.append(json.loads(line))
    return pd.DataFrame(data)

andrew_path = r"C:\BIANCONERI\Master's AI SJSU\5- Advanced Data Mining\Final Project\Yelp-JSON\Yelp JSON\yelp_dataset"
nic_path = r"C:\Users\nicor\OneDrive\Documents\SJSU\CMPE256\group_proj\Yelp JSON"

review_df = load_json(nic_path + r"\yelp_academic_dataset_review.json")
business_df = load_json(nic_path + r"\yelp_academic_dataset_business.json")
business_df = business_df[(business_df['state'] == 'CA') & (business_df['is_open'] == 1) & (business_df['categories'].notna())]
```

```python
business_df['categories'] = business_df['categories'].apply(lambda x: [category.lower() for category in x.split(', ')])
restaurant_keywords = ["bars","donuts","barbeque", "sandwiches","wineries","fish & chips","vegetarian", "beer", "food", "dessert", "gelato", "restaurants", "wine", "tacos", "tea", "acai bowls", "whiskey", "juice bars & smoothies", "poke", "spirits", "cocktail", "salad", "coffee", "bakeries", "breweries", "pizza", "burgers", "soup", "bagels", "ice cream & frozen yogurt", "ramen", "chicken wings", "food trucks", "cafes", "seafood", "vegan", "diners", "noodles", "comfort food", "soup", "asian fusion", "tex-mex", "dim sum", "smokehouse", "barbeque"]
restaurants_df = business_df[business_df['categories'].apply(lambda x: any(keyword in x for keyword in restaurant_keywords))]
```

```python
review_df = review_df[review_df['business_id'].isin(business_df['business_id'])]
```

```python
print(review_df['date'].min())
print(review_df['date'].max())
review_df['yyyy-mm'] = review_df['date'].str[0:7]
monthlies = pd.DataFrame()
monthlies['counts'] = review_df.groupby('yyyy-mm').size()
monthlies['cumulative'] = monthlies['counts'].cumsum()
# monthlies['year-mo'] = review_df['date'].str[0:7]

```

```python
monthlies.head()
monthlies.tail()
total = monthlies['cumulative'].iloc[-1]
splitpt = total * 0.8
print(splitpt)
monthlies['splitabs'] = abs(monthlies['cumulative']-splitpt)
splitmth = monthlies['splitabs'].argmin()
monthlies.iloc[splitmth]
```

```python
fig, axs = plt.subplots(2, figsize=(20,12))
_ = plt.subplots_adjust(hspace=0.3)
_ = axs[0].plot(monthlies['counts'])
_ = axs[0].set_xticklabels(review_df['yyyy-mm'],rotation=90)
_ = axs[0].axvline(x=splitmth,color='black')
_ = axs[0].set_title('Reviews per month')

_ = axs[1].plot(monthlies['cumulative'])
_ = axs[1].set_xticklabels(review_df['yyyy-mm'],rotation=90)
_ = axs[1].axvline(x=splitmth,color='black')
_ = axs[1].axhline(y = splitpt, color='red')
_ = axs[1].set_title('Cumulative total per month')
```
